4. Describe the architecture and functioning of a Recurrent Natural Network
(RNN).

       # Architecture and Functioning of Recurrent Neural Network (RNN)

"""
Recurrent Neural Network (RNN)
------------------------------
RNN is a type of neural network designed to handle sequential data
(such as text, speech, and time-series data).

------------------------------------------------------------

1. ARCHITECTURE OF RNN

RNN has a loop (recurrent connection) that allows information
to persist from previous steps.

Main components:
- Input layer (x_t)
- Hidden layer (h_t)
- Output layer (y_t)
- Recurrent connection (memory of previous hidden state)

Key idea:
h_t = function(x_t, h_(t-1))

This means current output depends on:
- Current input
- Previous hidden state (memory)

------------------------------------------------------------

2. WORKING / FUNCTIONING OF RNN

Step-by-step process:

Step 1: Input Sequence
Example: "I love AI"
Each word is given one-by-one:
x1 = "I"
x2 = "love"
x3 = "AI"

------------------------------------------------------------

Step 2: Hidden State Update
At each time step:

h_t = tanh(Wx * x_t + Wh * h_(t-1))

- x_t = current input
- h_(t-1) = previous memory
- h_t = updated memory

------------------------------------------------------------

Step 3: Output Generation
At each step or final step:

y_t = Wy * h_t

Output can be:
- Next word prediction
- Sentiment classification
- Sequence labeling

------------------------------------------------------------

3. KEY FEATURE (MEMORY)

- RNN remembers previous information using hidden state
- This makes it suitable for sequential data

------------------------------------------------------------

4. LIMITATIONS

- Vanishing gradient problem (difficulty learning long sequences)
- Short-term memory limitation
- Slow training for long sequences

------------------------------------------------------------

FINAL CONCLUSION:
RNN is a neural network that processes sequential data by maintaining
a memory (hidden state) of previous inputs, making it useful for NLP and time-series tasks.
"""

5. What is the key difference between LSTM and GRU networks in NLP
applications?
  
   # Key Difference between LSTM and GRU in NLP Applications

"""
LSTM (Long Short-Term Memory) and GRU (Gated Recurrent Unit)
are advanced types of Recurrent Neural Networks (RNNs)
used to solve the vanishing gradient problem.

------------------------------------------------------------

1. BASIC IDEA

LSTM:
- Uses 3 gates:
  1. Input gate
  2. Forget gate
  3. Output gate
- Has a separate cell state (memory) + hidden state

GRU:
- Uses 2 gates:
  1. Reset gate
  2. Update gate
- No separate cell state (simpler structure)

------------------------------------------------------------

2. ARCHITECTURE DIFFERENCE

LSTM:
- More complex architecture
- Maintains long-term and short-term memory separately
- More parameters

GRU:
- Simpler architecture
- Combines memory and hidden state
- Fewer parameters

------------------------------------------------------------

3. PERFORMANCE IN NLP

LSTM:
- Better for long sequences
- More accurate in complex tasks
- Slower training due to complexity

GRU:
- Faster training
- Works well on smaller datasets
- Slightly less powerful than LSTM in very long sequences

------------------------------------------------------------

4. COMPUTATIONAL COST

LSTM:
- Higher computation cost
- More memory usage

GRU:
- Lower computation cost
- More efficient

------------------------------------------------------------

FINAL CONCLUSION:

- LSTM → More powerful, better for long-term dependencies
- GRU → Faster, simpler, and more efficient alternative

Both are widely used in NLP tasks like:
- Machine translation
- Sentiment analysis
- Speech recognition
"""

8.  Implement a basic LSTM model in Keras for a text classification task using
the following dummy dataset. Your model should classify sentences as either positive
(1) or negative (0).
# Dataset
texts = [
“I love this project”, #Positive
“This is an amazing experience”, #Positive
“I hate waiting in line”, #Negative
“This is the worst service”, #Negative
“Absolutely fantastic!” #Positive ]
labels = [1, 1, 0, 0, 1]
Preprocess the text, tokenize it, pad sequences, and build an LSTM model to train on
this data. You may use Keras with TensorFlow backend.
(Include your Python code and output in the code box below.)


```python
# LSTM Text Classification with Preprocessing (Keras + TensorFlow)

import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# -------------------------
# Dataset
# -------------------------
texts = [
    "I love this project",
    "This is an amazing experience",
    "I hate waiting in line",
    "This is the worst service",
    "Absolutely fantastic!"
]

labels = [1, 1, 0, 0, 1]

# -------------------------
# 1. Tokenization
# -------------------------
tokenizer = Tokenizer(oov_token="<OOV>")
tokenizer.fit_on_texts(texts)

word_index = tokenizer.word_index

sequences = tokenizer.texts_to_sequences(texts)

# -------------------------
# 2. Padding
# -------------------------
max_len = max(len(seq) for seq in sequences)
padded_sequences = pad_sequences(sequences, maxlen=max_len, padding='post')

labels = np.array(labels)

# -------------------------
# 3. Build LSTM Model
# -------------------------
model = Sequential([
    Embedding(input_dim=len(word_index) + 1, output_dim=8, input_length=max_len),
    LSTM(16),
    Dense(1, activation='sigmoid')
])

# -------------------------
# 4. Compile Model
# -------------------------
model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

# -------------------------
# 5. Train Model
# -------------------------
model.fit(padded_sequences, labels, epochs=15, verbose=1)

# -------------------------
# 6. Evaluation
# -------------------------
loss, accuracy = model.evaluate(padded_sequences, labels, verbose=0)

print("Loss:", loss)
print("Accuracy:", accuracy)

# -------------------------
# 7. Prediction Example
# -------------------------
test_text = ["I really love this amazing service"]
test_seq = tokenizer.texts_to_sequences(test_text)
test_pad = pad_sequences(test_seq, maxlen=max_len, padding='post')

prediction = model.predict(test_pad)

print("Prediction Score:", prediction)
```

9.  Using spaCy, build a simple NLP pipeline that includes tokenization,
lemmatization, and entity recognition. Use the following paragraph as your dataset:
“Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the
development of India’s atomic energy program. He was the founding director of the Tata
Institute of Fundamental Research (TIFR) and was instrumental in establishing the
Atomic Energy Commission of India.”
Write a Python program that processes this text using spaCy, then prints tokens, their
lemmas, and any named entities found.
(Include your Python code and output in the code box below.)

```python
# spaCy NLP Pipeline: Tokenization, Lemmatization, and Named Entity Recognition

import spacy

# Load English model
nlp = spacy.load("en_core_web_sm")

# Input text
text = """Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the development of India’s atomic energy program.
He was the founding director of the Tata Institute of Fundamental Research (TIFR) and was instrumental in establishing the Atomic Energy Commission of India."""

# Process text
doc = nlp(text)

# -------------------------
# TOKENS + LEMMAS
# -------------------------
print("TOKENS AND LEMMAS:\n")
for token in doc:
    print(token.text, " ---> ", token.lemma_)

# -------------------------
# NAMED ENTITY RECOGNITION
# -------------------------
print("\nNAMED ENTITIES:\n")
for ent in doc.ents:
    print(ent.text, " ---> ", ent.label_)

"""
EXPECTED OUTPUT (approx):

TOKENS AND LEMMAS:
Homi ---> Homi
Jehangir ---> Jehangir
Bhaba ---> Bhaba
was ---> be
an ---> an
Indian ---> Indian
nuclear ---> nuclear
physicist ---> physicist
...

NAMED ENTITIES:
Homi Jehangir Bhaba ---> PERSON
Indian ---> NORP
India ---> GPE
Tata Institute of Fundamental Research ---> ORG
TIFR ---> ORG
Atomic Energy Commission of India ---> ORG
"""
```

10. You are working on a chatbot for a mental health platform. Explain how
you would leverage LSTM or GRU networks along with libraries like spaCy or Stanford
NLP to understand and respond to user input effectively. Detail your architecture, data
preprocessing pipeline, and any ethical considerations.
(Include your Python code and output in the code box below.)

```python
# Mental Health Chatbot Pipeline using spaCy + LSTM (Simple Demo)

import spacy
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# -------------------------
# SAMPLE DATA (intent classification)
# -------------------------
texts = [
    "I feel very sad and depressed",
    "I am stressed and anxious",
    "I am very happy today",
    "Life is beautiful and I feel great",
    "I am feeling lonely and hopeless",
    "Everything is going well"
]

labels = [2, 1, 0, 0, 2, 0]
# 0 = positive/normal
# 1 = stress/anxiety
# 2 = depression

# -------------------------
# TEXT PREPROCESSING USING SPACY
# -------------------------
def preprocess(text):
    doc = nlp(text.lower())
    tokens = [token.lemma_ for token in doc if token.is_alpha and not token.is_stop]
    return " ".join(tokens)

clean_texts = [preprocess(t) for t in texts]

# -------------------------
# TOKENIZATION
# -------------------------
tokenizer = Tokenizer()
tokenizer.fit_on_texts(clean_texts)

sequences = tokenizer.texts_to_sequences(clean_texts)

max_len = max(len(seq) for seq in sequences)
X = pad_sequences(sequences, maxlen=max_len, padding="post")

y = np.array(labels)

# -------------------------
# MODEL (LSTM)
# -------------------------
model = Sequential([
    Embedding(input_dim=len(tokenizer.word_index) + 1, output_dim=16),
    LSTM(16),
    Dense(3, activation="softmax")
])

model.compile(loss="sparse_categorical_crossentropy",
              optimizer="adam",
              metrics=["accuracy"])

# -------------------------
# TRAIN MODEL
# -------------------------
model.fit(X, y, epochs=20, verbose=1)

# -------------------------
# TEST PREDICTION
# -------------------------
test_text = "I am feeling very sad and stressed"
test_clean = preprocess(test_text)
test_seq = tokenizer.texts_to_sequences([test_clean])
test_pad = pad_sequences(test_seq, maxlen=max_len, padding="post")

prediction = model.predict(test_pad)

print("Prediction Probabilities:", prediction)

# -------------------------
# OUTPUT MEANING
# -------------------------
print("\nClass Meaning:")
print("0 = Positive/Normal")
print("1 = Stress/Anxiety")
print("2 = Depression")
```
